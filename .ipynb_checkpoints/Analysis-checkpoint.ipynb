### Using NetworkX

```python
import networkx as nx
```

```python
G = nx.Graph()
```

```python
NODES=list(set(tweet_df["UserID"]))
len(NODES)
```

```python
G.add_nodes_from(NODES)
```

```python
tweet_df.iloc[0]["Mentions"][1:-1].split(',')[0][1:-1]
```

```python
EDGES=[]

for i in range(tweet_df.shape[0]):
    temp_edges=[]
    tweeted_user=tweet_df.iloc[i]["UserID"]
    mentions=tweet_df.iloc[i]["Mentions"][1:-1].split(',')
    for m in mentions:
        temp_edges.append((tweeted_user,m[1:-1]))
    EDGES.extend(temp_edges)
```

```python
G.add_edges_from(EDGES)
```

```python
len(G.edges)
```

## Metrics

### Closeness Centrality

```python
closeness_centrality=nx.closeness_centrality(G)
```

```python
sorted_cc = sorted(closeness_centrality.items(), key = lambda kv: kv[1],reverse=True)
```

```python
sorted_cc[:10]
```

### Number of nodes and edges

```python
nodes_list=G.nodes
len(nodes_list)
```

```python
edges_list=G.edges
len(edges_list)
```

### Degree

```python
sorted(nx.degree(G),key=lambda x:x[1],reverse=True)
```

```python
sorted(nx.betweenness_centrality(G, normalized=True).items(), key=lambda x:x[1], reverse=True)[0:10]
```

```python
sorted(nx.closeness_centrality(G).items(), key=lambda x:x[1], reverse=True)[0:10]
```

```python
sorted(nx.eigenvector_centrality(G).items(), key=lambda x:x[1], reverse=True)[0:10]
```
